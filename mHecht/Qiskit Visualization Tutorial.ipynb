```python
import qiskit as qk
from qiskit import *
from qiskit.visualization import plot_histogram
from qiskit.tools.monitor import job_monitor
from qiskit import Aer
```

# Plot Histogram for 2qb Bell State

```python
bell = QuantumCircuit(2,2)
bell.h(0)
bell.cx(0,1)

bell.draw('mpl')
```

```python
measureqc = QuantumCircuit(2,2)
measureqc.measure([0,1],[0,1]) #measures quantum bit to classical bit
qc = bell + measureqc #combine measurement with bell circuit
qc.draw('mpl')
```

```python
backend  = BasicAer.get_backend('qasm_simulator')
result = execute(qc, backend, shots = 1024).result()
counts = result.get_counts(qc)
plot_histogram(counts)
```

## Comparison with Aer instead of BasicAer

```python
backend  = Aer.get_backend('qasm_simulator')
job = execute(qc, backend, shots = 1024)
results = job.result()
counts = results.get_counts(qc)
plot_histogram(counts)
```
