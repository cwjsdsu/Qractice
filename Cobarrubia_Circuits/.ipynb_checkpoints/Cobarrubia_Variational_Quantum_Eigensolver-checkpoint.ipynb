```python
import numpy as np
from qiskit import *
from sympy import *
from math import pi
from qiskit.circuit import Parameter
%matplotlib inline
```

```python
# Visualization packages
from qiskit.visualization import plot_state_city, plot_bloch_vector,plot_histogram
import matplotlib.pyplot as plt
```

```python
# Aer's qasm_simulator
backend_sim = Aer.get_backend('qasm_simulator')
backend_vec = Aer.get_backend('statevector_simulator')
backend_uni = Aer.get_backend('unitary_simulator')
```

```python
# Register circuit (total qubits in the network)
q_z = QuantumRegister(1) # quantum bit for z measurement
c_z = ClassicalRegister(1) # classical bit for z measurement
# Register circuit (total qubits in the network)
q_x = QuantumRegister(1) # quantum bit for x measurement
c_x = ClassicalRegister(1) # classical bit for x measurement
```

We want to parameterize some wave function |$\phi (x_i) \rangle$ such that we can the minimum eigenvalue of $\mathbf{A}$, $E_0$. This is because the Rayleigh-Ritz theorem for any wave function $| v \rangle$ is the following

$E_0 \le \frac{\langle v | \mathbf{A} | v \rangle}{\langle v | v \rangle}$.

We will consider a real, Hermitian matrix 
$\mathbf{A} = \begin{pmatrix} a_{11} & a_{12} \\ a_{21} & a_{22}\end{pmatrix} = \frac{a_{11} + a_{22}}{2}\mathbf{I} + a_{12}\mathbf{X} + \frac{a_{11}-a_{22}}{2}\mathbf{Z}$.

The average of $\mathbf{A}$ is simply $\langle \mathbf{A} \rangle =\frac{a_{11} + a_{22}}{2}\langle \mathbf{I} \rangle + a_{12} \langle \mathbf{X} \rangle + \frac{a_{11}-a_{22}}{2}\langle \mathbf{Z} \rangle$. This means we need to compute the $\langle \mathbf{X} \rangle$ and $\langle \mathbf{Z} \rangle$ first. 

We know that for a real, Hermitian matrix the eigenvectors will be real. We will propose that the parameterized wavefunction as
|$\phi (\theta) \rangle = \begin{pmatrix} cos(\theta) \\ sin(\theta)\end{pmatrix}$.

To make this, we need the qubit in the state $| 0 \rangle$ going to a rotational matrix $R(2\theta)$.



```python
# Parameters
theta_range = np.linspace(0,  np.pi, 250) # Theta linspace
total_shots = 10024 # Number of runs per trial of theta
```

## Measurement of $<Z>$


```python
#Measurement of <Z>
theta = Parameter('θ')
qc = QuantumCircuit(q_z,c_z)# Create a Quantum Circuit of a single qubit
initial_state = [1,0] # Inital State
qc.initialize(initial_state,0)# Initalize the quantum circuit 
qc.ry(2*theta,0)
qc.measure(q_z,c_z)

# Bind them to an array for the Backend to read
qc_circuits = [qc.bind_parameters({theta: theta_val})
              for theta_val in theta_range]

qc_circuits[-1].draw('mpl')
```

### Simulation  by qasm Backend

```python
# Single Measurement case
result_qc = execute(qc, backend_sim, shots=total_shots,memory = True).result() # get results from running the circuit
counts_qc = result_qc.get_counts(qc) # get counts 
memory_qc = result_qc.get_memory(qc) # all states collected during run
```

```python
# Getting results for a circuit over a parameter
results_z_theta = execute(qc,backend = backend_sim,shots = total_shots,parameter_binds = [{theta: theta_val} for theta_val in theta_range]).result()
counts_z_theta = results_z_theta.get_counts()
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)

ax.plot(theta_range, np.asarray(list(map(lambda c: c.get('0', 0), counts_z_theta)))/total_shots, '.-', label='0')
ax.plot(theta_range, np.asarray(list(map(lambda c: c.get('1', 0), counts_z_theta)))/total_shots, '.-', label='1')


ax.set_xticks([i * np.pi / 4 for i in range(5)])
ax.set_xticklabels(['0', r'$\frac{\pi}{4}$', r'$\frac{\pi}{2}$', r'$\frac{3\pi}{4}$', r'$\pi$'], fontsize=14)
ax.set_xlabel('θ', fontsize=14)
ax.set_ylabel('Probability', fontsize=14)
ax.legend(fontsize=14)
```

```python
Average_z =  np.subtract(np.asarray(list(map(lambda c: c.get('0', 0), counts_z_theta)))/total_shots, np.asarray(list(map(lambda c: c.get('1', 0), counts_z_theta)))/total_shots)
```

```python
fig = plt.figure(figsize=(8,6))
ax1 = fig.add_subplot(111)

ax1.plot(theta_range,Average_z, '.-', label='<Z>')



ax1.set_xticks([i * np.pi / 4 for i in range(5)])
ax1.set_xticklabels(['0', r'$\frac{\pi}{4}$', r'$\frac{\pi}{2}$', r'$\frac{3\pi}{4}$', r'$\pi$'], fontsize=14)
ax1.set_xlabel('θ', fontsize=14)
ax1.set_ylabel('$<\mathbf{Z}>$', fontsize=14)
ax1.legend(fontsize=14)
```

## Measurement of $<X>$

```python
#Measurement of <x>
theta = Parameter('θ')
qcx = QuantumCircuit(q_x,c_x)# Create a Quantum Circuit of a single qubit
initial_state = [1,0] # Inital State
qcx.initialize(initial_state,0)# Initalize the quantum circuit 
qcx.ry(2*theta,0)
qcx.h(0)
qcx.measure(q_x,c_x)

# Bind them to an array for the Backend to read
qcx_circuits = [qcx.bind_parameters({theta: theta_val})
              for theta_val in theta_range]

qcx_circuits[-1].draw('mpl')
```

```python
# Getting results for a circuit over a parameter
results_x_theta = execute(qcx,backend = backend_sim,shots = total_shots,parameter_binds = [{theta: theta_val} for theta_val in theta_range]).result()
counts_x_theta = results_x_theta.get_counts()
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)

ax.plot(theta_range, np.asarray(list(map(lambda c: c.get('0', 0), counts_x_theta)))/total_shots, '.-', label='0')
ax.plot(theta_range, np.asarray(list(map(lambda c: c.get('1', 0), counts_x_theta)))/total_shots, '.-', label='1')


ax.set_xticks([i * np.pi / 4 for i in range(5)])
ax.set_xticklabels(['0', r'$\frac{\pi}{4}$', r'$\frac{\pi}{2}$', r'$\frac{3\pi}{4}$', r'$\pi$'], fontsize=14)
ax.set_xlabel('θ', fontsize=14)
ax.set_ylabel('Probability', fontsize=14)
ax.legend(fontsize=14)
```

```python
Average_x =  np.subtract(np.asarray(list(map(lambda c: c.get('0', 0), counts_x_theta)))/total_shots, np.asarray(list(map(lambda c: c.get('1', 0), counts_x_theta)))/total_shots)
```

```python
fig = plt.figure(figsize=(8,6))
axx = fig.add_subplot(111)

axx.plot(theta_range,Average_x, '.-', label='<X>')



axx.set_xticks([i * np.pi / 4 for i in range(5)])
axx.set_xticklabels(['0', r'$\frac{\pi}{4}$', r'$\frac{\pi}{2}$', r'$\frac{3\pi}{4}$', r'$\pi$'], fontsize=14)
axx.set_xlabel('θ', fontsize=14)
axx.set_ylabel('$<\mathbf{X}>$', fontsize=14)
axx.legend(fontsize=14)
```

## Measurement of $<A>$

Now that we know the average of X and Z as a function of parameter $\theta$, we can then find the average of A. However, python can't do symbolic plotting (at least not in a trivial way), we will assume the following matrix
$\mathbf{A} = N \begin{pmatrix} 3 & \frac{3}{2} \\ \frac{3}{2} & 1\end{pmatrix}$,
where N is a normalization constant.

Therefore,
$\frac{\langle \mathbf{A}(\theta) \rangle}{N} = \frac{3 + 1}{2}\langle \mathbf{I}(\theta) \rangle + \frac{3}{2} \langle \mathbf{X}(\theta) \rangle + \frac{3-1}{2}\langle \mathbf{Z}(\theta) \rangle = 2 + \frac{3}{2} \langle \mathbf{X}(\theta) \rangle + \langle \mathbf{Z} (\theta)\rangle $

```python
# Measurement of <A>/N
avg_A1 = np.add(2,(3/2)*Average_x)
avg_A = np.add(avg_A1,Average_z)
```

```python
fig = plt.figure(figsize=(8,6))
axx = fig.add_subplot(111)

axx.plot(theta_range,avg_A, '.-', label='<A>')



axx.set_xticks([i * np.pi / 4 for i in range(5)])
axx.set_xticklabels(['0', r'$\frac{\pi}{4}$', r'$\frac{\pi}{2}$', r'$\frac{3\pi}{4}$', r'$\pi$'], fontsize=14)
axx.set_xlabel('θ', fontsize=14)
axx.set_ylabel('$<\mathbf{A}>/N$', fontsize=14)
axx.set_ylim(0,4)
axx.set_xlim(0,pi)
axx.legend(fontsize=14)
```

The groundstate of $<A>$ is when <A($\theta$)> is at a minimum. 

```python
np.min(avg_A) ## minimum A value
```

```python
theta_range[np.where(avg_A == np.min(avg_A))] # Find the theta when <a> is a minimum
```

According to this calculation, the groundstate of A is roughly
$E_0 \le 0.18N$ when $\theta = \frac{13\pi}{20}$

This means the ansatz wavefunction that will give the expectation of the lowest groundstate of $\mathbf{A}$ is
|$\phi \rangle = \begin{pmatrix} cos(\frac{13\pi}{20}) \\ sin(\frac{13\pi}{20})\end{pmatrix}$.
