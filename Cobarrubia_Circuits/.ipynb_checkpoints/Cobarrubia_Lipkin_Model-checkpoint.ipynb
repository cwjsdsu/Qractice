# Lipkin Model

```python
import numpy as np
from qiskit import *
from sympy import *
from math import pi
from qiskit.circuit import Parameter
%matplotlib inline
```

```python
# Visualization packages
from qiskit.visualization import plot_state_city, plot_bloch_vector,plot_histogram
import matplotlib.pyplot as plt
```

```python
# Aer's qasm_simulator
backend_sim = Aer.get_backend('qasm_simulator')
backend_vec = Aer.get_backend('statevector_simulator')
backend_uni = Aer.get_backend('unitary_simulator')
```

```python
# Register circuit (total qubits in the network)
q_1 = QuantumRegister(1) # quantum bit for z measurement
c_1 = ClassicalRegister(1) # classical bit for z measurement
# Register circuit (total qubits in the network)
q_2 = QuantumRegister(2) # quantum bit for x measurement
c_2 = ClassicalRegister(2) # classical bit for x measurement
```

```python
# Parameters
theta_range = np.linspace(0,  2*np.pi, 250) # Theta linspace
phi_range = np.linspace(0,  2*np.pi, 250) # phi linspace
total_shots = 10024
# Number of runs per trial of theta
```

Analytical Solution of $\langle H \rangle$ of the Lipkin Model:

$\langle H \rangle = -\epsilon J_z + \frac{V}{2}(J^2_+ + J^2_-)$ 

where N is number of "atoms", $\epsilon$ is related to energy, and $\chi = \frac{(N - 1)V}{\epsilon}$. $\langle H \rangle $ is at a minimum when  $\frac{1}{\chi} = cos(2\theta)$ or $\chi = 0$. We will evaluate the Lipkin model when N = 1 and $\epsilon = 1$.

# Part 1:

## Average of J_Z

```python
def jzcircuit(theta,initial_state, qubitregister,classicalregister):
    quantum_circuit = QuantumCircuit(qubitregister,classicalregister)# Create a Quantum Circuit of a single qubit
    quantum_circuit.initialize(initial_state,0)# Initalize the quantum circuit 
    quantum_circuit.ry(2*theta,0)
    quantum_circuit.measure(qubitregister,classicalregister)

###### OLD CODE #######
#theta = Parameter('Î¸')
#qcjz = QuantumCircuit(q_1,c_1)# Create a Quantum Circuit of a single qubit
#initial_state = [1,0] # Inital State
#qcjz.initialize(initial_state,0)# Initalize the quantum circuit 
#qcjz.ry(2*theta,0)
#qcjz.measure(q_1,c_1)
#
## Bind them to an array for the Backend to read
#qcjz_circuits = [qcjz.bind_parameters({theta: theta_val})
#              for theta_val in theta_range]
#
#qcjz_circuits[-1].draw('mpl')
    return quantum_circuit
```

```python
def exp_of_single_qb(list_of_quantumcircuits,theata_para,list_of_states,backend1,shots):
    #list_of_quantumcircuits = an array of the same quantum circuit over a single parameter (theta in this case)
    # theata_para = linspace of of theta
    # list_of_states = list of string of possible single-qubit states ['0','1']
    # Getting results for a circuit over a parameter
    
    results = execute(list_of_quantumcircuits,backend = backend1,shots = shots).result()
    counts = results.get_counts()
    
    Average = []
# For loop over all theta results in counts
    for theta_ind in range(len(counts)):
        probs = {} #initialize a probability array
        for output in list_of_states:
            if output in counts[theta_ind]:
                probs[output] = counts[theta_ind][output]/shots # if a count is a 0 or 1 then record the probability 
            else:
                probs[output] = 0
        Average.append((1/2)*(probs['0'] - probs['1'])) # attach average result for this theta slice
        
###### OLD CODE ######
# Getting results for a circuit over a parameter
#results_jz = execute(qcjz,backend = backend_sim,shots = total_shots,parameter_binds = [{theta: theta_val} for theta_val in theta_range]).result()
#counts_jz = results_jz.get_counts()
#Average_jz = []
## For loop over all theta results in counts
#for theta_ind in range(len(counts_jz)):
#    probs = {} #initialize a probability array
#    for output in ['0','1']:
#        if output in counts_jz[theta_ind]:
#            probs[output] = counts_jz[theta_ind][output]/total_shots # if a count is a 0 or 1 then record the probability 
#        else:
#            probs[output] = 0
#    Average_jz.append(probs['0'] - probs['1']) # attach average result for this theta slice
    return Average
```

```python
#Measurement of <J_Z>
init_state = [1,0]
qcjz = []
for index in range(len(theta_range)):
    qcjz.append(jzcircuit(theta_range[index],init_state, q_1,c_1))
    
    
Average_jz = exp_of_single_qb(qcjz,theta_range,['0','1'],backend_sim,total_shots)
    
analytical_jz = (1/2)*(np.cos(2*theta_range))
```

```python
fig = plt.figure(figsize=(8,6))
ax1 = fig.add_subplot(111)

ax1.plot(theta_range,Average_jz, '.-', label='Circuit Solution')
ax1.plot(theta_range,analytical_jz, '.-', label='Analytical Solution')


ax1.set_xticks([i * np.pi / 2 for i in range(5)])
ax1.set_xticklabels(['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'], fontsize=14)
ax1.set_xlabel('Î¸', fontsize=14)
ax1.set_ylabel('$<\mathbf{J_z}>$/N', fontsize=14)
ax1.legend(fontsize=14)
```

```python
#Measurement of < (J_+)^2 - (J_-)^2>
init_state1 = [1,0]
qcjpjm = []
for index in range(len(theta_range)):
    qcjpjm.append(JpJmcircuit(theta_range[index],init_state1,init_state1, q_2,c_2))
    
    
Average_jpjm = exp_of_JpJm(qcjpjm,theta_range,['00','01','10','11'],backend_sim,total_shots)
    
analytical_jpjm = (1/2)*(np.sin(2*theta_range))**2
```

```python
def H(V,eps,expJz, expJpJm):
    H = - eps*expJz - (V/2)*expJpJm
    return H
```

# Part 2: Generator Coordinate

Now we need to solve the off-diagonals of $J_z$ and $J^2_+ + J^2_-$ using an ancilla bit. The expectation of H is now

$\langle H \rangle/N = -\frac{\epsilon}{2}(cos(2\theta) + \frac{(N-1)V}{2\epsilon}sin^2(2 \theta))\frac{1 \pm cos^{N-1}(2\theta)}{1 \pm cos^N(2\theta)}$


```python
analytical_Hsmallchi = -(1/2)*(np.cos(2*theta_range) + 0.5*(np.sin(2*theta_range))**2)*((1 + np.cos(2*theta_range)**(-1))/(1 + np.cos(2*theta_range))) # chi < 1
analytical_Hbigchi = -(1/2)*(np.cos(2*theta_range) + 1.5*(np.sin(2*theta_range))**2)*((1 + np.cos(2*theta_range)**(-1))/(1 + np.cos(2*theta_range))) # chi > 1
```

```python
# Register circuit (total qubits in the network)
q_3 = QuantumRegister(2) # quantum bit for z measurement
c_3 = ClassicalRegister(2) # classical bit for z measurement
# Register circuit (total qubits in the network)
q_4 = QuantumRegister(3) # quantum bit for x measurement
c_4 = ClassicalRegister(3) # classical bit for x measurement
```

```python
def JpJmoffcircuit(theta,initial_state_qb1,initial_state_qb2, initial_state_qb3, qubitregister,classicalregister):
    quantum_circuit = QuantumCircuit(qubitregister,classicalregister)# Create a Quantum Circuit of a single qubit
    quantum_circuit.initialize(initial_state_qb1,0)# Initalize the quantum circuit 
    quantum_circuit.initialize(initial_state_qb2,1)# Initalize the quantum circuit 
    quantum_circuit.initialize(initial_state_qb3,2)# Initalize the quantum circuit 
    quantum_circuit.h(0) # Apply the Hadamard gate
    quantum_circuit.ry(-2*theta,1) # Apply Rotational-Y matrix on qb 2
    quantum_circuit.ry(-2*theta,2) # Apply Rotational-Y matrix on qb 3
    quantum_circuit.cx(0,1) # Apply Rotational-Y matrix on qb 2
    quantum_circuit.ry(2*theta,1) # Apply CNOT gate controlling qb 1 and targeting qb2
    quantum_circuit.cx(0,1) # Apply Rotational-Y matrix on qb 2
    quantum_circuit.ry(2*theta,1) # Apply CNOT gate controlling qb 1 and targeting qb2
    quantum_circuit.cx(0,2) # Apply Rotational-Y matrix on qb 3
    quantum_circuit.ry(2*theta,2) # Apply CNOT gate controlling qb 1 and targeting qb3
    quantum_circuit.cx(0,2) # Apply Rotational-Y matrix on qb 3
    quantum_circuit.ry(2*theta,2) # Apply CNOT gate controlling qb 1 and targeting qb3
    quantum_circuit.cx(1,2)# Apply CNOT gate controlling qb 2 and targeting qb3
    quantum_circuit.barrier()
    quantum_circuit.h(0) # Apply the Hadamard gate
    quantum_circuit.h(1) # Apply the Hadamard gate
    quantum_circuit.barrier()
    quantum_circuit.measure(range(3),range(3))
    

    return quantum_circuit
```

```python
def exp_of_JpJmoff(list_of_quantumcircuits,theata_para,list_of_states,backend1,shots):
    #list_of_quantumcircuits = an array of the same quantum circuit over a single parameter (theta in this case)
    # theata_para = linspace of of theta
    # list_of_states = list of string of possible single-qubit states ['0','1']
    # Getting results for a circuit over a parameter
    
    results = execute(list_of_quantumcircuits,backend = backend1,shots = shots).result()
    counts = results.get_counts()
    
    Average = []
# For loop over all theta results in counts
    for theta_ind in range(len(counts)):
        probs = {} #initialize a probability array
        
        for output in list_of_states:
            if output in counts[theta_ind]:
                probs[output] = counts[theta_ind][output]/shots # if a count is a 0 or 1 then record the probability 
            else:
                probs[output] = 0
        Average.append((probs['000'] - probs['110'] - probs['001'] + probs['111'])) # attach average result for this theta slice
        

    return Average
```

```python
#Measurement of off-diagonals < -theta|J^2_+ - J^2_-|theta> 
init_state = [1,0]
circuit_list_offJpJm = []
# Loop over all theta angles
for index_1 in  range(len(theta_range)):
    circuit_list_offJpJm.append(JpJmoffcircuit(theta_range[index_1],init_state,init_state,init_state, q_4,c_4))

Average_offjpjm = exp_of_JpJmoff(circuit_list_offJpJm,theta_range,['000','001','010','011','101','100','110','111'],backend_sim,total_shots)
ana_off_jpjm = (np.cos(2*theta_range)**(-1))*np.sin(2*theta_range)**2
```
