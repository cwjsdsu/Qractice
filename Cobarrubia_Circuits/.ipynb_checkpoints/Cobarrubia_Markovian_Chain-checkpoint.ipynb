## Markovian Chain

Following a conference lecture "Markov process simulation on a real
quantum computer" $[$ AIP Conference Proceedings 2172 $]$ by Nikolov et. al., I will attempt to code their Markovian chain.

A Markovian chain is simply the propbability of one state transitioning to another state,  where the state i + 1 only depends on the current iteration i. Each transition is "memoryless" and "causal" such that $t_0 < t_1<...<t_N$. Here they propose that this transition is the entanglement of multiple qubits where entanglement is created a n-sqrt(X) matrix.

Given a N-qubit system, for each qubit they will represent a singular time step i. For a state to have a probabilitic depdency on the current state, the n-sqrt(X) is controlled to the previous qubit. For example, a 2-qubit system will be

q_0 ----- GATE$[$n-sqrt(X)$]$ ------    CONTROLLELD     ------------


q_1 --------------------------------TARGET$[$ n-sqrt(X)$]$ ---------

Only the first and last qubit, the state won't be control another qubit. Also, the reason why we have a n-sqrt(X) gate in the first qubit is because the state won't have a probabilisitc transition to another timestep. Therefore, we need to push it to have a transition state. 


Here we can represent any quantum state  as $|\psi \rangle = \sum_i a_i |i\rangle = e^{i\phi}cos(\theta) |0\rangle + sin(\theta)|1\rangle$. To represent a time step, they propose that n-sqrt(X)$|0\rangle$ will give you the same behavior. This gate would require a phase gate to create. 

To create the  n-sqrt(X) we know that


 $$ n-sqrt(PHASE) =\begin{pmatrix} 1 & 0 \\ 0 & e^{i\pi/n} \end{pmatrix} $$.
 
 This is also known as the n-sqrt(Z) gate. This is jsut he square root matrix of the Z-gate, and the Hadamard operator transforms the Z-gate to the X-gate. Therefore, we can make a unitary transformation such that 
 
  $ n-sqrt(X) = \mathbf{H}$ (n-sqrt(PHASE)) $\mathbf{H} $
  
  or
  
  GATE$[n-sqrt(X)]$ = GATE$[H]$---GATE$[n-sqrt(PHASE)]$ --- GATE$[H]$
  


NOTE: There is an error in their lecture. The plots say $\alpha_0 = 0.25$ and $\alpha_1 = 0.75$ when they should be $\alpha^2_0$ and $\alpha^2_1$.

```python
def nroot_control_gate(qc,phi,control,qb):
    qc.h(qb)
    qc.crz(phi,control,qb)
    qc.h(qb)
    return qc
```

```python
def nroot_gate(qc,phi,qb):
    qc.h(qb)
    qc.rz(phi,qb)
    qc.h(qb)
    return qc
```

```python
def Three_step_circuit(phi,initial_state,qr,cr):
    quantum_circuit = QuantumCircuit(qr,cr)
    quantum_circuit.initialize(initial_state,0)# Initalize the quantum circuit 
    quantum_circuit.initialize(initial_state,1)# Initalize the quantum circuit 
    quantum_circuit.initialize(initial_state,2)# Initalize the quantum circuit 
    quantum_circuit = nroot_gate(quantum_circuit,phi,0)
    quantum_circuit = nroot_gate(quantum_circuit,0,1)
    quantum_circuit = nroot_gate(quantum_circuit,0,2)
    quantum_circuit.barrier()
    quantum_circuit = nroot_control_gate(quantum_circuit,phi,0,1)
    quantum_circuit.barrier()
    quantum_circuit = nroot_control_gate(quantum_circuit,phi,1,2)
    quantum_circuit.barrier()
    #quantum_circuit.measure_all()
    quantum_circuit.measure(range(3),range(3))
    return quantum_circuit
```

```python
def Five_step_circuit(phi,initial_state,qr,cr):
    quantum_circuit = QuantumCircuit(qr,cr)
    quantum_circuit.initialize(initial_state,0)# Initalize the quantum circuit 
    quantum_circuit.initialize(initial_state,1)# Initalize the quantum circuit 
    quantum_circuit.initialize(initial_state,2)# Initalize the quantum circuit 
    quantum_circuit.initialize(initial_state,3)# Initalize the quantum circuit 
    quantum_circuit.initialize(initial_state,4)# Initalize the quantum circuit 
    quantum_circuit = nroot_gate(quantum_circuit,phi,0)
    quantum_circuit = nroot_gate(quantum_circuit,0,1)
    quantum_circuit = nroot_gate(quantum_circuit,0,2)
    quantum_circuit = nroot_gate(quantum_circuit,0,3)
    quantum_circuit = nroot_gate(quantum_circuit,0,4)
    quantum_circuit.barrier()
    quantum_circuit = nroot_control_gate(quantum_circuit,phi,0,1)
    quantum_circuit.barrier()
    quantum_circuit = nroot_control_gate(quantum_circuit,phi,1,2)
    quantum_circuit.barrier()
    quantum_circuit = nroot_control_gate(quantum_circuit,phi,2,3)
    quantum_circuit.barrier()
    quantum_circuit = nroot_control_gate(quantum_circuit,phi,3,4)
    quantum_circuit.barrier()
    #quantum_circuit.measure_all()
    quantum_circuit.measure(range(5),range(5))
    return quantum_circuit
```

```python
qc_5 = Five_step_circuit(pi/(1/(np.arccos((2*0.25-1))/pi)),[1,0],q_1,c_1)
qc_5.draw('mpl')
```

```python
results = execute(qc_5,backend = backend_sim,shots = N_shots).result()
counts = results.get_counts()
plot_histogram(counts)
```

I believe, according to their description, that these probabilities will make the transition matrix, $\Pi$. Since there are N-qubits representing a step with $2^N$ possible paths. The transition matrix is a 2 $\times$ 2 matrix. To find the transition matrix from step 1 to the last step then

$$ \Pi = \mathbf{\Pi}_{i=1}^{N-1}\Pi_i $$
where the transition matrix for step n is $\Pi_n$.

For the 3-qb circuit the each transition matrix represent qb 1 state = 0 or 1 $->$ qb 2 state = 0 or 1 $->$ qb 3 state = 0 or 1.

Here the transition matrix for the 3-step qubit circuit from step 1 to step 3 is

$$ \mathbf{\Pi} =  \begin{pmatrix} P(000) + P(010)  & P(001) + P(011) \\ P(100) + P(110) & P(101) + P(111) \end{pmatrix} = \begin{pmatrix} 25.3 \% & 33.2 \% \\ 0 & 41.5 \% \end{pmatrix}. $$

The probability of P(total state) is really the probability of path P(state of qubit 1 $->$ state of qubit 2 $->$ state of qubit 3). For example, P($|001\rangle$) = P(1 $->$ 0 $->$ 0).

Notice that the bottom left of the matrix is zero. This means there is no path from the first qubit state = 0 to the last qubit state = 1. Here the sum of all components equals 1 which satisfies the condition to be a transition matrix.
